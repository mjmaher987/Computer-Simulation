#Scan the Input
In this part we are going to scan the input.

```python
n = int(input())
intial_probs = list(map(float, input().split()))
transition_probs = []
for i in range(n):
  row = list(map(float, input().split()))
  transition_probs.append(row)
```

```python
print("Initial probs are: ")
print(intial_probs)
```

```python
print("transition probs are: ")
print(transition_probs)
```

#6-a) Find Steady State Distribution


image.png

```python
import random
all_states = [i for i in range(n)]
steady_states = []
simulation_number = 20000 # You can change it
for s_number in range(simulation_number):
  state = random.choices(all_states, intial_probs)[0]
  steps = 17 # You can change it
  for step in range(steps):
    new_probs = transition_probs[state]
    state = random.choices(all_states, new_probs)[0]
  steady_states.append(state)
```

```python
print(steady_states)
```

```python
for i in range(n):
  print("Probabilty of being in state " + str(i) +" is " + str(steady_states.count(i) / len(steady_states)))
```

```python
for i in range(n):
  answer = round(steady_states.count(i) / len(steady_states), 2)
  print(str(answer), end=' ')
```

##Effect of Steps

```python
import random
all_states = [i for i in range(n)]
all_probs = []
for steps in range(1, 40):
  steady_states = []
  simulation_number = 20000 # You can change it
  for s_number in range(simulation_number):
    state = random.choices(all_states, intial_probs)[0]
    for step in range(steps):
      new_probs = transition_probs[state]
      state = random.choices(all_states, new_probs)[0]
    steady_states.append(state)
  this_prob = []
  for i in range(n):
    answer = round(steady_states.count(i) / len(steady_states), 2)
    this_prob.append(answer)
  all_probs.append(this_prob)
print(all_probs)
```

##Effect of Simulation Number

```python
import random
all_states = [i for i in range(n)]
all_probs = []
steps = 17
for simulation_number in range(1000, 20000, 2000):
  steady_states = []
  for s_number in range(simulation_number):
    state = random.choices(all_states, intial_probs)[0]
    for step in range(steps):
      new_probs = transition_probs[state]
      state = random.choices(all_states, new_probs)[0]
    steady_states.append(state)
  this_prob = []
  for i in range(n):
    answer = round(steady_states.count(i) / len(steady_states), 2)
    this_prob.append(answer)
  all_probs.append(this_prob)
print(all_probs)
```

```python
import seaborn as sns
sns.heatmap(all_probs,cmap='magma',linecolor='white',linewidths=1)
```
